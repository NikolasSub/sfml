```python
import pandas as pd
```

```python
train = pd.read_csv('03_santander_train.csv', error_bad_lines=False, low_memory=False, index_col='ID')
```

```python
train.head()
```

```python
test = pd.read_csv('03_santander_test.csv', index_col='ID')
```

```python
test.head()
```

```python
y = list(train['TARGET'])
```

```python
train = train.drop('TARGET', axis=1)
```

```python
train.info()
```

```python
all_data = pd.concat([train, test], ignore_index=True)
```

```python
all_data.info(verbose=True, memory_usage=True, null_counts=True)
```

```python
all_data.isnull().any()
```

```python
all_data = all_data.fillna(all_data.median())
```

```python
y = pd.DataFrame(y)
```

```python
y = y.fillna(y.median())
```

```python
y.isnull().any()
```

```python
data_train = all_data.loc[:y.shape[0]-1]
data_test = all_data.loc[y.shape[0]:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_train, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
#Нужно взять sklearn.tree.DecisionTreeClassifier 
#Визуализировать как меняется ROC_AUC от max_depth
predict_probs = []
for depth in range(1, 15):
    dtc = DecisionTreeClassifier(max_depth=depth, random_state = 42)
    dtc.fit(X_train, y_train)
    predict = dtc.predict(X_test)
    predict_proba = dtc.predict_proba(X_test)
    predict_probs.append((predict_proba, depth))
plt.figure(figsize=(5, 10))
plt.xlabel('tpr')
plt.ylabel('fpr')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)
for element in predict_probs:
    predict_proba_dtc, depth_dtc = element
    score = roc_auc_score(y_test, predict_proba_dtc[:,1])
    fpr, tpr, thresholds = roc_curve(y_test, predict_proba_dtc[:,1])
    plt.plot(fpr, tpr, label = f'ROC_dtc score {score:.3f} {depth_dtc}')
    plt.legend(borderaxespad = 0.)
```

```python
#Нужно взять sklearn.neighbors.KNeighborsClassifier 
#Визуализировать как меняется ROC_AUC от n_neighbors (данный метод долго считает predict потому можно уменьшить 
#кол-во фичей по которым учится модель и считаются данные)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
predict_probs = []
for neighbors in range(1, 15):
    knr = KNeighborsClassifier(n_neighbors=neighbors, n_jobs = 4)
    knr.fit(X_train.iloc[:20000, :150], y_train.iloc[:20000])
    predict = knr.predict(X_test.iloc[:20000, :150])
    predict_proba = knr.predict_proba(X_test.iloc[:20000, :150])
    predict_probs.append((predict_proba, neighbors))
plt.figure(figsize=(5, 10))
plt.xlabel('tpr')
plt.ylabel('fpr')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)
for element in predict_probs:
    predict_proba_knr, neighbor_knr = element
    score = roc_auc_score(y_test.iloc[:20000], predict_proba_knr[:,1])
    fpr, tpr, thresholds = roc_curve(y_test.iloc[:20000], predict_proba_knr[:,1])
    plt.plot(fpr, tpr, label = f'ROC_knr score {score:.3f} {neighbor_knr}')
    plt.legend(borderaxespad = 0.)
```

```python
#Нужно взять sklearn.linear_model.LogisticRegression 
#Визуализировать как меняется ROC_AUC от параметров penalty и C
```

```python
predict_probs = []
for c in range(1, 15):
    lr = LogisticRegression(penalty ='l1', C = c)
    lr.fit(X_train, y_train)
    predict = lr.predict(X_test)
    predict_proba = lr.predict_proba(X_test)
    predict_probs.append((predict_proba, c))
plt.figure(figsize=(5, 10))
plt.xlabel('tpr')
plt.ylabel('fpr')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)
for element in predict_probs:
    predict_proba_lr, c_lr = element
    score = roc_auc_score(y_test, predict_proba_lr[:,1])
    fpr, tpr, thresholds = roc_curve(y_test, predict_proba_lr[:,1])
    plt.plot(fpr, tpr, label = f'ROC_lr score {score:.3f} {c_lr}')
    plt.legend(borderaxespad = 0.)
```
